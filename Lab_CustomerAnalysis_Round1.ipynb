```python
#Question 1
import pandas as pd
import numpy as np

# Reading data into a DataFrame
file1 = pd.read_csv("/Users/devirughani/Desktop/IronHack/Week_1/Day_2/Labs/lab-customer-analysis-round-1/files_for_lab/csv_files/file1.csv")
file2 = pd.read_csv("/Users/devirughani/Desktop/IronHack/Week_1/Day_2/Labs/lab-customer-analysis-round-1/files_for_lab/csv_files/file2.csv")
file3 = pd.read_csv("/Users/devirughani/Desktop/IronHack/Week_1/Day_2/Labs/lab-customer-analysis-round-1/files_for_lab/csv_files/file3.csv")
```

```python
#Show the DataFrame's shape.

print(file1.shape)
print(file2.shape)
print(file3.shape)
```

```python
#Standardize header names

display(file1.head(1))
display(file2.head(1))
display(file3.head(1))
```

```python
file1= file1.rename(columns={'ST':'State',
                            'GENDER':'Gender'})
file2=file2.rename(columns={'ST':'State',
                            'GENDER':'Gender'})

display(file1.head(1))
display(file2.head(1))
display(file3.head(1))
```

```python
file1=file1[['Customer', 'State','Gender',  'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Total Claim Amount', 'Vehicle Class']]
file2=file2[['Customer', 'State','Gender',  'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Total Claim Amount', 'Vehicle Class']]
file3=file3[['Customer', 'State','Gender',  'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Total Claim Amount', 'Vehicle Class']]

display(file1.head(1))
display(file2.head(1))
display(file3.head(1))
```

```python
#Merge/concatenate the 3 files
data = pd.concat([file1, file2, file3], axis=0)
data.shape
```

```python
data.dtypes
#Which columns are numerical? Income, Monthly Premium Auto and Total Claim amount are currently numerical

#Which columns are categorical? State, Gender, Policy Type and Vehicle Class
```

Delete the column education and the number of open complaints from the dataframe.

```python
data = data.drop(['Education','Number of Open Complaints'], axis=1)
display(data.head())
```

Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.

```python
#Dropped duplicates
print(data.shape)
data2=data.copy().drop_duplicates()
print(data2.shape)
```

```python
data2['Customer Lifetime Value'] =  pd.to_numeric(data2['Customer Lifetime Value'], errors='coerce')
print(data2.dtypes)
print(data2.shape)
```

```python
data2.head()
display(data2['Customer Lifetime Value'])
```

```python
def clvx100(x):
    return x*100

data2['Customer Lifetime Value'] = list(map(clvx100, data2['Customer Lifetime Value']))
display(data2['Customer Lifetime Value'])
```
